```python
%%R
mpg01 <- ifelse(Auto$mpg > median(Auto$mpg), 1, 0)
dat <- data.frame(subset(Auto, select=-c(mpg, name)), mpg01)
dat$origin <- as.factor(dat$origin)
```

```python
%%R
pairs(dat, col=dat$mpg01 + 2)
```

```python
%%R
svm.fit <- svm(mpg01~horsepower + acceleration, kernel="linear", data=dat, cost=10, scale=TRUE)
```

```python
%%R
svm.pred <- predict(svm.fit, data=dat, type="class")
```

```python
%%R
plot(svm.pred)
```
