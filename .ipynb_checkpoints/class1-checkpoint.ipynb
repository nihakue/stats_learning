```python
%load_ext rmagic
```

```python
%%R
n = 100
ngrid = 1000
nrep = 100
sigma = 0.1
mym = 3
myd = 0.1

myfun = function( x, m, d ){
  y = m*x
  y[x>0.5] = m - m*x[x>0.5]
  x1 = 0.5-d
  x2 = 0.5+d
  y[x<x1] = m*x1
  y[x>x2] = m*x1
  y
}

x = runif(n)
# to make it easy later on, choose the first and last values of x to be 0 and 1
x[1] = 0
x[n] = 1
epsilon = rnorm( n, 0, sigma )
y = myfun( x, mym, myd ) + epsilon

# compute sensible plotting limits
ymin = myfun( 0, mym, myd ) - 5*sigma
ymax = myfun( 0.5, mym, myd ) + 5*sigma
plot( x, y, xlim=c(0,1), ylim=c(ymin,ymax), pch=20, col="red" )

# draw the "true" function
x0 = seq( from=min(x), to=max(x), length=ngrid)
truef = myfun(x0,mym,myd)
lines( x0, truef, type="l", col="blue", lwd=2)

# draw the interpolated solution (low bias, high variance)
interp.out = approx( x, y, xout=x0 )
pred1 = interp.out$y
lines( x0, pred1, col="orange", lwd=2 )


# draw the linear regression solution (high bias, low variance)
lm.out = lm( y~x )
pred2 = predict( lm.out, newdata=data.frame( x=x0 ) )
lines( x0, pred2, col="cyan", lwd=2 )

# draw the loess-smoothed solution (moderate bias, moderate variance)
loess.out = loess( y~x, span=0.5 )
pred3 = predict( loess.out, x0 )
lines( x0, pred3, col="magenta", lwd=2 )
```

```python
%%R

# now do many reps
pred1 = matrix( 0, ngrid, nrep )
pred2 = matrix( 0, ngrid, nrep )
pred3 = matrix( 0, ngrid, nrep )

plot( x, y, xlim=c(0,1), ylim=c(ymin,ymax), type="n" ) # empty plot
lines( x0, truef, type="l", col="blue", lwd=2)
for( i in 1:nrep ){
  x = runif(n)
  x[1] = 0
  x[n] = 1
  epsilon = rnorm( n, 0, sigma )
  y = myfun( x, mym, myd ) + epsilon
  # draw the "true" function
  x0 = seq( from=min(x), to=max(x), length=ngrid)
  truef = myfun(x0,mym,myd)
  lines( x0, truef, type="l", col="blue", lwd=2)
  # draw the interpolated solution (low bias, high variance)
  interp.out = approx( x, y, xout=x0 )
  pred1[,i] = interp.out$y
  lines( x0, pred1[,i], col="orange", lwd=2 )
  # draw the linear regression solution (high bias, low variance)
  lm.out = lm( y~x )
  pred2[,i] = predict( lm.out, newdata=data.frame( x=x0 ) ) 
  lines( x0, pred2[,i], col="cyan", lwd=2 )
  # draw the loess-smoothed solution (moderate bias, moderate variance)
  loess.out = loess( y~x, span=0.5 )
  pred3[,i] = predict( loess.out, x0 )
  lines( x0, pred3[,i], col="magenta", lwd=2 )
}
```

```python
%%R
bias1 = apply( pred1, 1, mean ) - truef
bias1sq = bias1^2
var1 = apply( pred1, 1, var )

bias2 = apply( pred2, 1, mean ) - truef
bias2sq = bias2^2
var2 = apply( pred2, 1, var )

bias3 = apply( pred3, 1, mean ) - truef
bias3sq = bias3^2
var3 = apply( pred3, 1, var )

par(mfrow=c(3,2)) 
plot( x0, bias1, type="l", col="orange", lwd=2, ylab="bias" )
plot( x0, var1, type="l", col="orange", lwd=2, ylab="variance" )
plot( x0, bias2, type="l", col="cyan", lwd=2, ylab="bias" )
plot( x0, var2, type="l", col="cyan", lwd=2, ylab="variance" )
plot( x0, bias3, type="l", col="magenta", lwd=2, ylab="bias" )
plot( x0, var3, type="l", col="magenta", lwd=2, ylab="variance" )

par(mfrow=c(3,2)) 
myylimb = c(0, max(c(bias1sq,bias2sq,bias3sq)))
myylimv= c(0, max(c(var1,var2,var3)))
plot( x0, bias1sq, type="l", col="orange", lwd=2, ylim=myylimb, ylab="squared bias" )
legend("topright", paste("mean = ",signif(mean(bias1sq),2)))
plot( x0, var1, type="l", col="orange", lwd=2, ylim=myylimv, ylab="variance" )
legend("topright", paste("mean = ",signif(mean(var1),2)))
plot( x0, bias2sq, type="l", col="cyan", lwd=2, ylim=myylimb, ylab="squared bias" )
legend("topright", paste("mean = ",signif(mean(bias2sq),2)))
plot( x0, var2, type="l", col="cyan", lwd=2, ylim=myylimv, ylab="variance" )
legend("topright", paste("mean = ",signif(mean(var2),2)))
plot( x0, bias3sq, type="l", col="magenta", lwd=2, ylim=myylimb, ylab="squared bias" )
legend("topright", paste("mean = ",signif(mean(bias3sq),2)))
plot( x0, var3, type="l", col="magenta", lwd=2, ylim=myylimv, ylab="variance" )
legend("topright", paste("mean = ",signif(mean(var3),2)))

par(mfrow=c(1,1))
e1 = bias1sq+var1
e2 = bias2sq+var2
e3 = bias3sq+var3
plot( x0, e1, type="l", col="orange", lwd=2, ylim=c(0,max(c(e1,e2,e3))), ylab = "variance + squared bias" )
lines( x0, e2, type="l", col="cyan", lwd=2 )
lines( x0, e3, type="l", col="magenta", lwd=2 )

legend("topright", 
  c(paste("mean = ",signif(mean(e1),2)),
    paste("mean = ",signif(mean(e2),2)),
    paste("mean = ",signif(mean(e3),2))),
  lty=c(1,1,1), lwd=c(2,2,2), col = c("orange","cyan","magenta"))
abline(h=0)
```
